# V. MERGING AND JOINING

## A. MERGE

#### 1. Merge is used to merge two data frame into one data frame

```python
import pandas as pd
df1 = pd.DataFrame({
    "Name" : ["Guru", "Vishnu", "Bala"],
    "Age" : ["9", "45", "19"]
})
df1
```

```python
df2 = pd.DataFrame({
    "Name" : ["Bala", "Guru", "Vishnu"],
    "Life" : ["College", "School", "Work"]
})
df2
```

```python
Data5 = pd.merge(df1,df2,on="Name")
Data5
```

#### 2. How= "inner" is used to print commom data in both data frame

```python
import pandas as pd
df3 = pd.DataFrame({
    "Name" : ["Guru", "Vishnu", "Kowsic", "Jeeva"],
    "Age" : ["9", "45", "17","67"]
})
df1
```

```python
df4 = pd.DataFrame({
    "Name" : ["Bala", "Guru", "Vishnu"],
    "Life" : ["College", "School", "Work"]
})
df4
```

```python
Data6 = pd.merge(df3,df4,on="Name", how="inner")
Data6
```

#### 3. How= "outer" is used to print all data in both data frame

```python
Data7 = pd.merge(df3,df4,on="Name", how="outer")
Data7
```

#### 4. How= "left" is used to print only first dataframe

```python
Data8 = pd.merge(df3,df4,on="Name", how="left")
Data8
```

#### 5. How= "left" is used to print only last dataframe


```python
Data9 = pd.merge(df3,df4,on="Name", how="right")
Data9
```

#### 6. "indicator= True" is used to find location of the row data by adding a new column

```python
import pandas as pd
Data10 = pd.merge(df3,df4,on="Name", how= "outer", indicator=True)
Data10
```

# VI. TIME SERIES ANALYSIS

## A. HANDLING DATE AND TIME DATA

```python
import pandas as pd
Guru0 = pd.read_csv('DailyDelhiClimateTrain.csv', parse_dates=['date'])
Guru0.head()
```

#### 1. This is used to convert object into date 

```python
import pandas as pd
Guru0['date'] = pd.to_datetime(Guru0['date'])
Guru0.head()
```

#### 2. This is used to Separate Date, Month, Month Name, Days, WeekDays, Year, Current Date,etc..,

```python
Guru0['Date'] = Guru0.date.dt.date
Guru0['Month'] = Guru0.date.dt.month
Guru0['MonthName'] = Guru0.date.dt.month_name()
Guru0['Days'] = Guru0.date.dt.day
Guru0['WeekDays'] = Guru0.date.dt.day_name()
Guru0['Year'] = Guru0.date.dt.year
Guru0['Current Date'] = pd.Timestamp.today()

Guru0.head()
```

#### 3. This is used to add/ subtract Days, Month, Year From a Date

```python
import pandas as pd
from dateutil.relativedelta import relativedelta
Guru0['Day Before'] = Guru0['Date'] - relativedelta(days=10)
Guru0.head()
```

```python
import pandas as pd
from dateutil.relativedelta import relativedelta
Guru0['Day After'] = Guru0['Date'] + relativedelta(days=10)
Guru0.head()
```

#### 4. This is used to print the month after/before as end/begin

```python
import pandas as pd
from dateutil.relativedelta import relativedelta
Guru0['Month End'] = Guru0['Date'] + pd.offsets.MonthEnd()
Guru0.head()
```
